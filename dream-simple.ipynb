```python
# Setup
import dream
```

```python
# Change settings here
model_path = '../caffe/models/bvlc_googlenet/bvlc_googlenet.caffemodel'
image_path = 'images/adam.jpg'
iterations = 10
noctaves = 4
octave_scale = 1.4
layer = 'inception_4c/output'
```

```python
# Dream
net = dream.load_net(model_path)
img = dream.load_image(image_path)
print 'Starting dream...'
dream_img = dream.deepdream(net, img, iterations, noctaves, octave_scale, layer)
print 'Done dreaming'
dream.showarray(dream_img)
```

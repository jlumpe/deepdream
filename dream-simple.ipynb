```python
# Import stuff
import image
import dream
```

```python
# Import the net
model_path = '../caffe/models/bvlc_googlenet/bvlc_googlenet.caffemodel'
net = dream.load_net(model_path)
```

```python
# Load / create the image

# Load from file
image_path = 'images/scream.jpg'
img = image.load(image_path)
#img = image.resize(image, height=480) # Resize to 480p

# Random noise
#img = image.noise(480)

# Perlin noise
#img = image.perlin(9, 1)
```

```python
# Dream

# Settings
iterations = 20
noctaves = 4
octave_scale = 1.4
layer = 'inception_4c/output'

# Can re-run this cell to continue dreaming on previous output
print 'Starting dream...'
img = dream.deepdream(net, img, iterations, noctaves, octave_scale, layer)
print 'Done dreaming'
image.show(img)
```

```python
# Run this to save the image
image.save(img, 'dreams/my-dream.jpg')
```
